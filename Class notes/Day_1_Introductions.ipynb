# What is biophysics?
Biophysics represents a science of the fundamental physical laws underlying the structure, functioning, and development of living systems.

Screen Shot 2024-03-14 at 1.04.26 PM.jpg

# How big are the systems that biophysics studies?
Biophysics covers various organization levels, starting with biomacromolecules and then at the cellular and subcellular level, at the level of organs, organisms, populations and communities, ecosystems, and finally, at the level of the biosphere as a whole.


# How do we study these systems in biophysics
Through experiments and modeling. In this class, we will focus on the modeling aspect. Basic models represented by one or two (or more) equations allowing a qualitative examination, make it possible to describe  regularities of biological processes: growth processes and restrictions, presence of several stable stationary states, oscillations, quasistochastic regimes, travelling pulses and waves, and the structures inhomogeneous in space.

This type of models includes the models of growth, interaction between species, patterns that microorganisms make, intracellular calcium oscillations, glycolysis, nerve conductivity, and DNA untwisting, to name just a few.
